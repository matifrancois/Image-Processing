Como el enunciado nos solicita que permitamos al usuario modificar el valor correspondiente al tamaño del elemento estructural, a continuación se muestra el diseño de una interfaz de usuario que implementa un slicer que permite cambiar el número de pixeles que contempla el elemento estructural y un botón para imprimir la imágen según el valor de dicho número. Para eso se hace uso de la librería ipywidgets.

De todas maneras se deja a continuación una imagen de la interfaz en caso de que no se pueda visualizar correctamente.

```python
import ipywidgets as widgets
from IPython.display import display
from IPython.display import clear_output
import matplotlib.pyplot as plt
from ipywidgets import HBox, VBox


int_range = widgets.IntSlider(min = 0, max = 100, value = 30)
output = widgets.Output()

display(int_range, output)

Int = 30

def on_value_change(change):
    with output:
        clear_output(wait=True)
        global Int
        Int = change['new']

int_range.observe(on_value_change, names='value')

def on_button1_clicked(b):
    with out:
        global Int 
        clear_output(wait=True)
        kernel = np.ones((Int,Int),np.uint8)
        opened = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
        fig=plt.figure(figsize=(14, 10))
        fig.add_subplot(1, 2, 1)
        plt.title("d")
        plt.imshow(img, cmap='gray')
        plt.plot()
        fig.add_subplot(1, 2, 2)
        plt.title("m")
        plt.imshow(opened, cmap='gray')
        plt.plot()
        plt.show()

button1 = widgets.Button(description='Procesar')

button1.on_click(on_button1_clicked)
out = widgets.Output()

left_box2 = VBox([button1])

HBox([left_box2, out])
```
