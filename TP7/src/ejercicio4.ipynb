Extraer el contorno de la figura suministrada por la cátedra.

A continuación se importan las librerías que se utilizarán.

```python
import cv2
import numpy as np
from helper_file import plot_images
```

A continuación se lee la imágen.

```python
img = cv2.imread('../assets/Lincoln from penny.tif', cv2.IMREAD_GRAYSCALE)
```

Se crea el kernel que se utilizará para el procesado.

```python
kernel =  cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
print(kernel)
```

Se opera la imágen con el kernel generado y se muestran los resultados en la imágen.

```python
img_contour = cv2.erode(img, kernel, iterations=1)
plot_images(img, 'original', img - img_contour, 'contour by erosion')
```
