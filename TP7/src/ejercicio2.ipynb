A continuación se desarrolla una interfaz de usuario que permite modificar el tamaño del kernel, el mismo es un parámetro muy útil ya que según su tamaño será el nivel de detalle que elimina sobre la imágen original. 

La interfaz de usuario que implementa un slicer que permite cambiar el número de pixeles que contempla el elemento estructural y un botón para imprimir la imágen según el valor de dicho número. Para eso se hace uso de la librería ipywidgets.

```python
import ipywidgets as widgets
from IPython.display import display
from IPython.display import clear_output
import matplotlib.pyplot as plt
from ipywidgets import HBox, VBox


int_range = widgets.IntSlider(min = 1, max = 22, value = 11)
output = widgets.Output()

display(int_range, output)

Int = 11

def on_value_change(change):
    with output:
        clear_output(wait=True)
        global Int
        Int = change['new']

int_range.observe(on_value_change, names='value')

def on_button1_clicked(b):
    with out:
        global Int 
        clear_output(wait=True)
        kernel = np.ones((Int,Int),np.uint8)
        opened = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
        fig=plt.figure(figsize=(14, 10))
        fig.add_subplot(1, 2, 1)
        plt.title("d")
        plt.imshow(img, cmap='gray')
        plt.plot()
        fig.add_subplot(1, 2, 2)
        plt.title("m")
        plt.imshow(opened, cmap='gray')
        plt.plot()
        plt.show()

button1 = widgets.Button(description='Procesar')

button1.on_click(on_button1_clicked)
out = widgets.Output()

left_box2 = VBox([button1])

HBox([left_box2, out])
```
