# Ejercicio 1

Aplicar la operación de dilación a la imagen del texto que se provee, siendo el elemento 
estructural:

$$
B = \begin{bmatrix}
0 & 1 & 0\\
1 & 1 & 1\\
0 & 1 & 0
\end{bmatrix}
$$

A continuación importamos las bibliotecas que utilizaremos.

```python
import cv2
import numpy as np
from helper_file import plot_images
```

Leemos la imágen a utilizar.

```python
img = cv2.imread('../assets/Broken_Text.tif', cv2.IMREAD_GRAYSCALE)
```

Generamos el kernel según lo que solicita la consigna

```python
kernel =  cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
print(kernel)
```

Operamos e imprimimos el resultado.

```python
img_dilated = cv2.dilate(img, kernel, iterations=1)
plot_images(img, 'original', img_dilated, 'dilatada')
```

Podemos ver que los trazos de las líneas que conforman las letras ahora son mas gruesos y que se puede leer con menores problemas el texto ya que las líneas negras que aparecen en las letras como errores tienden a minimizarse.
