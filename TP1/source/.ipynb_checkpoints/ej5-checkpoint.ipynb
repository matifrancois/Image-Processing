# Interpolación Bicúbica

```python
# Padding funcion (Monochromatic)

def image_padding(img, padding_size=2, padding_same=True):
    # Get image height and width
    h, w = img.shape
    ps = padding_size
    
    # Create new image with padding (adding 2*ps rows and 2*ps cols)
    pad_img = np.zeros(h+2*ps, w+2*ps)
    
    # Copy image, centered
    pad_img[ps:h+ps, ps:w+ps] = img

    
    if padding_same == True :
        # Extending image rows/cols to padding
        
        # Padding top
        pad_img[0:ps, ps:w+ps] = img[0, :]
        
        # Padding bottom
        pad_img[h+ps:h+2*ps, ps:w+ps] = img[h, :]
        
        # Padding right
        pad_img[ps:h+ps, w+ps:w+2*ps] = img[:, w]
        
        # Padding left
        pad_img[ps:h+ps, 0:ps] = img[:, 0]
        
        # Padding top-right corner
        pad_img[0:ps, w+ps:w+2*ps] = img[0, w]
        
        # Padding bottom-right corner
        pad_img[h+ps:h+2*ps, w+ps:w+2*ps] = img[h, w]
        
        # Padding bottom-left corner
        pad_img[h+ps:h+2*ps, 0:ps] = img[h, 0]
        
        # Padding top-left corner
        pad_img[0:ps, 0:ps] = img[0,0]
        
    return pad_img


# Bicubic function
# Returns the bicubic 1D weight of a given pixel. 
def bic_w(s, a=-0.5):

    x = np.abs(s)
    
    if (x < 1):
        w = (a+2)*(x**3) - (a+3)*(x**2) + 1
        
    else if (x >= 1) and (x < 2):
        w = a*(x**3) - 5*a*(x**2) + 8*a*x - 4*a
    
    else:
        w = 0
        
    return w
    
# Bicubic interpolation function
# Scales an image using bicubic interpolation. Scale ratio must be an integer

def bicubic_interpolation(img, ratio=2, a=-0.5, padding_same=True):
    k = int(ratio)
    dh, dw = img.shape * k
    
    # Destination image padding
    dest = image_padding(img=img, padding_size=2, padding_same=padding_same)
    
    # Iterate over every pixel in destination image
    for i in range(dh):
        for j in range(dw):
            
            # Scale large image coords to fit source ones (+2 padding offset)
            x = (j / k) + 2
            y = (i / k) + 2
            
            # Get adjacent columns coords (in small image)
            x1 = int(np.floor(x) - 1)
            x2 = x1 + 1
            x3 = x1 + 2
            x4 = x1 + 3
            
            # Get adjacent rows coords (in small image)
            y1 = int(np.floor(y) - 1)
            y2 = y1 + 1
            y3 = y1 + 2
            y4 = y1 + 3
            
            # Get weight matrix
            A = np.matrix([[bic_w(x-x1, a), bic_w(x-x2, a), bic_w(x-x3, a), bic_w(x-x4, a)]])
            
            # Get pixel value matrix
            B = np.matrix([
                            [img[y1,x1], img[y2,x1], img[y3,x1], img[y4,x1]],
                            [img[y1,x2], img[y2,x2], img[y3,x2], img[y4,x2]],
                            [img[y1,x3], img[y2,x3], img[y3,x3], img[y4,x3]],
                            [img[y1,x4], img[y2,x4], img[y3,x4], img[y4,x4]]
                
                        ])
            
            C = np.matrix([
                            [bic_w(y-y1, a)],
                            [bic_w(y-y2, a)],
                            [bic_w(y-y3, a)],
                            [bic_w(y-y4, a)]
                
                        ])
            
            dest[i, j] = np.dot(np.dot(A,B),C)
            
    return dest
```
